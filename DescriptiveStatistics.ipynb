```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/계통한계가격과 수도권 자급률 연구/MetroConcentration.csv')
df
```

```python
# SMP 연도별 추이 확인

# 숫자화 + 연도 기준으로 오름차순 정렬
df["Year"] = pd.to_numeric(df["Year"])
df["SMP"]  = pd.to_numeric(df["SMP"])
df["LNG price"]  = pd.to_numeric(df["LNG price"])
df = df.sort_values("Year")

# 그래프 출력
plt.plot(df["Year"], df["SMP"], marker="o")
plt.plot(df["Year"], df["LNG price"], marker="o")
plt.xlabel("Year")
plt.ylabel("SMP")
plt.title("SMP & LNG Price")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(["SMP", "LNG Price"])
plt.show()
```

```python
# 수도권 에너지 자급률 연도별 추이 확인

# 숫자화 + 연도 기준으로 오름차순 정렬
df["Year"] = pd.to_numeric(df["Year"])
df["SudoEnegrySelf"]  = pd.to_numeric(df["SudoEnegrySelf"])
df = df.sort_values("Year")

# 그래프 출력
plt.plot(df["Year"], df["SudoEnegrySelf"], marker="o")
plt.xlabel("Year")
plt.ylabel("SudoEnergySelf")
plt.title("SudoEnergySelf")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(["SudoEnergySelf"])
plt.show()
```
